**Function to convert the answer of each question into numerical labels, to feed to NLP model later.**

```python
def letter_to_label(letter):
    if 'A' <= letter <= 'Z':
        return ord(letter) - ord('A')
    else:
        return None  # Handle invalid input (not an uppercase letter)
letter_to_label('E')
```

**Making a new column "label" to store numerical label of the answer.**

```python
train_df["label"] = train_df["answer"].map(letter_to_label)
train_df
```

**The AutoTokenizer class in this case will download the pretrained BertModel tokenizer to our working environment.**

```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained(os.path.join("/kaggle/input", "roberta-base"))
```

**This function will concatenate the question with both the five options for each question in the training dataset, and then using the tokenizer loaded above to tokenized the resulting sentence from that concatenation.**

```python
options = ["A", "B", "C", "D", "E"]


def multiple_choice_question_preprocessing_function(example):
    first_sentences = [ example["prompt"] for i in range(5) ]
    second_sentences = [
        example[option] for option in options
    ]

    tokenized_examples = tokenizer(first_sentences, second_sentences, truncation=True)
    return {k: v for k, v in tokenized_examples.items()}
```

**Mapping the function to the dataset and we will have the tokenized training dataset.**

```python
tokenized_train_dataset = train_dataset.map(multiple_choice_question_preprocessing_function)
```

```python
from dataclasses import dataclass
from transformers.tokenization_utils_base import PreTrainedTokenizerBase, PaddingStrategy
from typing import Optional, Union
import tensorflow as tf


@dataclass
class DataCollatorForMultipleChoice:
    """
    Data collator that will dynamically pad the inputs for multiple choice received.
    """

    tokenizer: PreTrainedTokenizerBase
    padding: Union[bool, str, PaddingStrategy] = True
    max_length: Optional[int] = None
    pad_to_multiple_of: Optional[int] = None

    def __call__(self, features):
        label_name = "label" if "label" in features[0].keys() else "labels"
        labels = [feature.pop(label_name) for feature in features]
        batch_size = len(features)
        num_choices = len(features[0]["input_ids"])
        flattened_features = [
            [{k: v[i] for k, v in feature.items()} for i in range(num_choices)] for feature in features
        ]
        flattened_features = sum(flattened_features, [])

        batch = self.tokenizer.pad(
            flattened_features,
            padding=self.padding,
            max_length=self.max_length,
            pad_to_multiple_of=self.pad_to_multiple_of,
            return_tensors="tf",
        )

        batch = {k: tf.reshape(v, (batch_size, num_choices, -1)) for k, v in batch.items()}
        batch["labels"] = tf.convert_to_tensor(labels, dtype=tf.int64)
        return batch
```

```python
def cross_validation_training(cv = 5):

    cv_accuracy = []

    index_list = [i for i in range(len(tokenized_train_dataset))]
    random.shuffle(index_list)
    index_lists = np.array_split(index_list, cv)

    for cv_phase in range(cv):
        batch_size = 8
        num_train_epochs = 15
        total_train_steps = (len(tokenized_train_dataset) // batch_size) * num_train_epochs
        optimizer, schedule = create_optimizer(init_lr = 1e-5, num_warmup_steps = 0, num_train_steps = total_train_steps)

        model = TFAutoModelForMultipleChoice.from_pretrained("roberta-base")

        data_collator = DataCollatorForMultipleChoice(tokenizer = tokenizer)
        tf_train_set = model.prepare_tf_dataset(
            Dataset.from_dict(tokenized_train_dataset[ np.setdiff1d(np.array(index_list), index_lists[cv_phase])]),
            shuffle = True,
            batch_size = batch_size,
            collate_fn = data_collator,
        )

        tf_validation_set = model.prepare_tf_dataset(
            Dataset.from_dict(tokenized_train_dataset[index_lists[cv_phase]]),
            shuffle = False,
            batch_size = batch_size,
            collate_fn = data_collator,
        )

        model.compile(optimizer = optimizer, metrics = ["accuracy"])

        #metric_callback = KerasMetricCallback(metric_fn = compute_metrics, eval_dataset = tf_validation_set)
        model.fit(x = tf_train_set, validation_data = tf_validation_set, epochs = 15)

        accuracy_metrics = tf.keras.metrics.Accuracy()
        for instance, label in tf_validation_set.as_numpy_iterator():
            prediction = model.predict(instance).logits
            prediction = tf.argmax(prediction, axis = 1)
            accuracy_metrics.update_state(label, prediction)

        cv_accuracy.append(accuracy_metrics.result().numpy())

    return cv_accuracy
```

```python
from dataclasses import dataclass
from transformers.tokenization_utils_base import PreTrainedTokenizerBase, PaddingStrategy
from typing import Optional, Union
import tensorflow as tf


@dataclass
class DataCollatorForMultipleChoiceInference:
    """
    Data collator that will dynamically pad the inputs for multiple choice received.
    """

    tokenizer: PreTrainedTokenizerBase
    padding: Union[bool, str, PaddingStrategy] = True
    max_length: Optional[int] = None
    pad_to_multiple_of: Optional[int] = None

    def __call__(self, features):
        batch_size = len(features)
        num_choices = len(features[0]["input_ids"])
        flattened_features = [
            [{k: v[i] for k, v in feature.items()} for i in range(num_choices)] for feature in features
        ]
        flattened_features = sum(flattened_features, [])

        batch = self.tokenizer.pad(
            flattened_features,
            padding=self.padding,
            max_length=self.max_length,
            pad_to_multiple_of=self.pad_to_multiple_of,
            return_tensors="tf",
        )

        batch = {k: tf.reshape(v, (batch_size, num_choices, -1)) for k, v in batch.items()}
        return batch
```

```python
batch_size = 8

infer_data_collator = DataCollatorForMultipleChoiceInference(tokenizer = tokenizer)
tf_test_set = model.prepare_tf_dataset(
    tokenized_test_dataset,
    shuffle = False,
    batch_size = batch_size,
    collate_fn = infer_data_collator,
)
```

```python
test_top_k_values, test_top_k_indices = tf.math.top_k(test_logit_prediction, k = 3, sorted = True)
test_top_k_indices
```

```python
def convert_label_to_letter(top_k_array):
    letter_array = []
    def label_to_letter(label):
        return chr(label + ord('A'))
    for num_option in top_k_array:
        letter_array.append(label_to_letter(num_option))
    return tf.constant(letter_array)
```

```python
test_top_k_predictions = tf.map_fn(fn = lambda t: convert_label_to_letter(t), elems = test_top_k_indices, dtype = tf.string)
test_top_k_predictions
```

```python
# open the file in the write mode
with open(CSV_TEST_RESULT_DIRECTORY, 'w', newline = '') as f:
    # create the csv writer
    writer = csv.writer(f)
    header = ["id", "prediction"]
    writer.writerow(header)

    for index, top_k_prediction in enumerate(list(test_top_k_predictions)):
        row = [ index ]
        prediction_string = []
        for prediction in top_k_prediction:
            prediction_string.append(prediction.numpy().decode('utf-8'))
        row.append(" ".join(prediction_string))
        writer.writerow(row)
```
